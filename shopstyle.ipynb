```python
import time
import re
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
s = Service()
chrome_options = webdriver.ChromeOptions()
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])                                 # keep automation
chrome_options.add_experimental_option('useAutomationExtension', False)                                # bot hidden on web
chrome_options.add_experimental_option("detach", True)
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
# chrome_options.add_argument("--incognito")                                                     # To keep THE browser active
prefs = {"profile.default_content_setting_values.notifications" : 2}
chrome_options.add_experimental_option("prefs", prefs)
driver = webdriver.Chrome(chrome_options=chrome_options, service= s)
from bs4 import BeautifulSoup
driver.implicitly_wait(5)
driver.maximize_window()
```

```python
links = []
for item in driver.find_elements(By.XPATH,'//span[@class="products-list__item products-list__cell ng-star-inserted"]'):
    link = item.find_element(By.XPATH,'//a[@class="product-cell__view-details"]').get_attribute('href')
    cluster_link = item.find_element(By.XPATH,'//a[@data-test="product-cell__link"]').get_attribute('href')
    product_name = item.find_element(By.XPATH,'//span[@class="ss-t-text-ellipsis ss-w-full"]').text.strip()
    branded_product_name = item.find_element(By.XPATH,'//span[@data-test="product-cell__product-name"]').text.replace('at','').strip()
#     try:
    msrp = item.find_element(By.XPATH,'//span[@class="product-cell__price product-cell__price--struckout"]').text.replace('$','').strip()
    sale_price = item.find_element(By.XPATH,'//span[@class = "product-cell__price ss-red ng-star-inserted"]').text.replace('$','').strip()
    discount = str(int(100 - (float(sale_price)/float(msrp)*100)))
#     except:
#         sale_price = item.find_element(By.XPATH,'//span[@data-test="product-cell__price"]').text.strip()
#         msrp = ''
#         discount = ''
    data_dict = {"Link":link, 'Cluster Link':cluster_link, 'Product Name':product_name,'Branded Product Name':branded_product_name, "MSRB":msrp, 'Sale Price':sale_price, 'Discount':discount}
    print(data_dict)
    links.append(data_dict)
```

```python
links = []
for item in driver.find_elements(By.XPATH, '//span[@class="products-list__item products-list__cell ng-star-inserted"]'):
    link = item.find_element(By.XPATH, './/a[@class="product-cell__view-details"]').get_attribute('href')
    cluster_link = item.find_element(By.XPATH, './/a[@data-test="product-cell__link"]').get_attribute('href')
    product_name = item.find_element(By.XPATH, './/span[@class="ss-t-text-ellipsis ss-w-full"]').text.strip()
    branded_product_name = item.find_element(By.XPATH, './/span[@data-test="product-cell__product-name"]').text.strip()
    retailer_name = item.find_element(By.XPATH, './/span[@class="product-cell__at-retailer ss-t-center ss-he-full ss-w-full"]').text.replace('at','').strip()
    try:
        msrp = item.find_element(By.XPATH, './/span[@class="product-cell__price product-cell__price--struckout"]').text
        msrp = msrp.replace('$', '').strip()
        sale_price = item.find_element(By.XPATH, './/span[@class="product-cell__price ss-red ng-star-inserted"]').text
        sale_price = sale_price.replace('$', '').strip()
        discount = str(int(100 - (float(sale_price) / float(msrp) * 100)))
    except:
        sale_price = item.find_element(By.XPATH, './/span[@data-test="product-cell__price"]').text
        sale_price = sale_price.replace('$', '').strip()
        msrp = ''
        discount = ''
    
    data_dict = {
        "Link": link,
        'Cluster Link': cluster_link,
        'Product Name': product_name,
        'Branded Product Name': branded_product_name,
        "MSRP": msrp,
        'Sale Price': sale_price,
        'Discount': discount,
        'Retailer Name': retailer_name
    }
    print(data_dict)
    links.append(data_dict)
```

```python
df =pd.DataFrame(links)
df.head()
```

```python
driver = webdriver.Chrome(chrome_options=chrome_options, service= s)
for i,row in df.iterrows():
    url = row['Cluster Link']
    try:
        driver.get(url)
    except:
        pass
    
    time.sleep(2)
    
    try:
        df.loc[i, 'BrandName'] = driver.find_element(By.XPATH,'//a[@data-test="brand-title__link"]').text.strip()
    except:
        pass
    
    try:
        driver.find_element(By.XPATH,'//div[@class="product-page__col-description-caret"]').click()
    except:
        pass
    
    try:
        df.loc[i, 'Image Url'] = driver.find_element(By.XPATH,'//img[@websize="XLarge"]').get_attribute('src')
    except:
        pass
    
    try:
        df.loc[i, 'Details'] = driver.find_element(By.XPATH,'//p[@data-test="product-infos__description-text"]').text.strip().replace('\n','')
    except:
        pass
    
    print(df.loc[i, 'Details'])
    
    
```
